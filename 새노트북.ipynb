```python
def draw(way, random_color = False):
    for i, w in enumerate(way):
        xy = np.array(np.array(w['node']))
        if random_color:
            plt.plot(xy[:,0],xy[:,1])
            continue
        if w['tag']['type'] == "stop_line":
            plt.plot(xy[:,0],xy[:,1], color='red', zorder=2, linewidth = 4, label = 'stop_line')
        elif w['tag']['type'] == "pedestrian_marking":
            plt.plot(xy[:,0],xy[:,1], color='skyblue', zorder=1, linewidth = 4, label = 'pedestrian_marking')
        elif w['tag']['type'] == "traffic_sign":
            plt.plot(xy[:,0],xy[:,1], color='blue', zorder=1, linewidth = 5, label = 'traffic_sign')
        elif w['tag']['type'] == "virtual":
            plt.plot(xy[:,0],xy[:,1], color='black', zorder=1, label = 'virtual')
        elif w['tag']['type'] == "curbstone":
            plt.plot(xy[:,0],xy[:,1], color='orange', zorder=1, label = 'curbstone')
        elif w['tag']['type'] == "line_thin":
            plt.plot(xy[:,0],xy[:,1], color='gray', zorder=1, linewidth = 1, label = 'line_thin')
        elif w['tag']['type'] == "line_thick":
            plt.plot(xy[:,0],xy[:,1], color='gray', zorder=1, linewidth = 4, label = 'line_thick')
        elif w['tag']['type'] == "road_border":
            plt.plot(xy[:,0],xy[:,1], color='yellow', zorder=10, linewidth = 4, label = 'road_border')
        else:
            plt.plot(xy[:,0],xy[:,1], color='gray', label = 'others')
    plt.legend()
    plt.show()
```
